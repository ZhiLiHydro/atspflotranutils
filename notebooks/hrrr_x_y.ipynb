```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.loadtxt('./xyz_csv/x.csv', delimiter=',')
y = np.loadtxt('./xyz_csv/y.csv', delimiter=',')
```

```python
x.shape, y.shape
```

```python
x.flatten().shape[0], y.flatten().shape[0]
```

```python
xyz = np.zeros((x.flatten().shape[0], 3))
for i in range(x.flatten().shape[0]):
    xyz[i, 0], xyz[i, 1] = x.flatten()[i], y.flatten()[i]
```

```python
plt.plot(xyz[:, 0], xyz[:, 1], '.')
plt.show()
```

```python
np.savetxt('./xyz_csv/xyz.xyz', xyz)
```

```python
xyz_daymet = np.loadtxt('./xyz_csv/xyz_daymet.xyz')
```

```python
x_daymet = xyz_daymet[:, 0].reshape(x.shape)
y_daymet = xyz_daymet[:, 1].reshape(y.shape)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].plot(xyz_daymet[:, 0], xyz_daymet[:, 1], linestyle='none', marker='.', color='silver')
axs[1].plot(x_daymet, y_daymet, linestyle='none', marker='.', color='silver')
axs[1].plot(x_daymet[0, 0], y_daymet[0, 0], '.', c='red')
axs[1].plot(x_daymet[-1, -1], y_daymet[-1, -1], '.', c='blue')
axs[1].plot(x_daymet[-1, 0], y_daymet[-1, 0], '.', c='green')
axs[1].plot(x_daymet[0, -1], y_daymet[0, -1], '.', c='orange')
plt.show()
```

```python
xv, yv = np.meshgrid(np.linspace(x_daymet.min(), x_daymet.max(), 40), np.linspace(y_daymet.min(), y_daymet.max(), 29))
xv.shape, yv.shape
```

```python
plt.figure(figsize=(12, 12))
plt.plot(x_daymet, y_daymet, marker='.', lw=2, linestyle='none', zorder=2)
plt.plot(xv, yv, marker='.', color='k', linestyle='none', zorder=4)
plt.grid(lw=5, color='lightgrey', zorder=1)
plt.axis('equal')
plt.title('Colored points: HRRR coordinates converted to Daymet coordinates (irregular grid)\nBlack points: what ATS needs (regular grid)')
plt.show()
```
